```python
import math

# TASK 1
def sumList(numbers):
    # Returns the sum of all the values in a list
    total_sum = sum(numbers)
    return total_sum

 
def minList(numbers):
    # Returns the minimum value in a list
    return min(numbers)

 

 
    # min_num = numbers[0]
    # for i in range(0, len(numbers)):    
    #     #Compare elements of array with min    
    #     if(numbers[i] < min_num):    
    #         min_num = numbers[i]
     
    # return min_num
   
def maxList(numbers):
    # Returns the maximum value in a list
    return min(numbers)


 
    # max_num = numbers[0]
    # for i in range(0, len(numbers)):    
    #     #Compare elements of array with min    
    #     if(numbers[i] > max_num):    
    #         max_num = numbers[i]
   
    # return max_num

 
def rangeList(numbers):
    # Returns the range of values in a list
    range = max(numbers) - min(numbers)
    return range

 
def muList(numbers):
    # Returns the average of the values in a list
    mu = sum(numbers)/len(numbers)
    return mu

 
def medianList(numbers):
    # Returns the median of the values in a list
   
    numbers.sort()
    mid = len(numbers) // 2
    median = (numbers[mid] + numbers[~mid]) / 2
    return median

 
def modeList(numbers):
    # Returns the mode of the values in a list
    mode = 0
    temp = {}
    for num in numbers:
        if not num in temp:
            temp[num] = 1
        else:
            temp[num] += 1
    mode = [key for key,value in temp.items() if value == max(temp.values())][0]        
    return mode

 
def sigma2List(numbers):
    # Returns the Population variance deviation
    mu = muList(numbers)
    sigma2 = 0
    for num in numbers:
        sigma2 += (num-mu)**2 # sum of square of (x-mu)
   
    sigma2 = sigma2/len(numbers) # divided by N
    return sigma2

 
def sigmaList(numbers):
   
    # Returns the Population standard deviation
    mu = muList(numbers)
    sigma = 0
    for num in numbers:
        sigma += (num-mu)**2 # sum of square of (x-mu)
   
    sigma = sigma/(len(numbers)) # divided by N-1
    sigma = math.sqrt(sigma)
    return sigma

 
def sampleStdDeviation(numbers):
   
    # Returns the Population standard deviation
    mu = muList(numbers)
    sigma = 0
    for num in numbers:
        sigma += (num-mu)**2 # sum of square of (x-mu)
   
    sigma = sigma/(len(numbers) -1) # divided by N-1
    sigma = math.sqrt(sigma)
    return sigma


 
def sampleVarianceList(numbers):
   
    # Returns the Population standard deviation
    mu = muList(numbers)
    sigma = 0
    for num in numbers:
        sigma += (num-mu)**2 # sum of square of (x-mu)
   
    sigma = sigma/(len(numbers) -1) # divided by N-1
    sigma = math.sqrt(sigma)
    return sigma
   

 
def percentileList(numbers, k): 
    # Returns the specific percentile of the values in a list
    max_num = max(numbers)
    temp = (k/100) * max_num
   
    percentile = math.ceil(temp)
    return percentile


 #TASK-2
def descriptiveStatSummary():
   
    numbers = [1,1,2,3,3,3,3,4,4,1,2,2]
    print(f"Data Values: {len(numbers)}")
    print(f"Min Value: {minList(numbers)}")
    print(f"Max Value: {maxList(numbers)}")
    print(f"Range: {rangeList(numbers)}")
    print(f"Mean Value: {muList(numbers)}")
    print(f"Mode: {modeList(numbers)}")

 
    print(f"Polulation Variance: {sigma2List(numbers)}")
    print(f"Sample Variance: {sampleVarianceList(numbers)}")
    print(f"Polulation Standard Deviation: {sigmaList(numbers)}")
    print(f"Sample Standard Deviation: {sampleStdDeviation(numbers)}")

 
     # other
    print(f"Median Value: {medianList(numbers)}")
    print(f"Percentile of 2nd : {percentileList(numbers, 2)}")

 
if __name__ == "__main__":
    descriptiveStatSummary()
```

```python
#TASK:4
import math
def addVectors(list1 , list2):
    minimum_length = min(len(list1), len(list2))
    i = 0
    result = []
    while i<minimum_length:
        result.append(list1[i] + list2[i])
        i += 1
    if i < len(list1) :
        result += list1[i:len(list1)]
    elif i < len(list2) :
        result += list2[i:len(list2)]

    return result

def subtractVectors(list1  , list2):
    minimum_length = min(len(list1), len(list2))
    i = 0
    result = []
    while i<minimum_length:
        result.append(list1[i] - list2[i])
        i += 1
    if i < len(list1) :
        result -= list1[i:len(list1)]
    elif i < len(list2) :
        result -= list2[i:len(list2)]

    return result

def scaleVector(list1 , scalar_value):
    return [(scalar_value * d ) for d in list1]

def normalizeVector(list1):
    squares = [ math.pow(value , 2) for value in list1]
    D = math.sqrt(sum(squares)) 
    return [ value/D for value in list1]

def variance(data):
   
  n = len(data)
   
  mean = sum(data) / n
   
  deviations = [(x - mean) ** 2 for x in data]
    
  variance = sum(deviations) / n
  return variance

def mean(data):
  n = len(data)
  mean = sum(data) / n
  return mean

def z_score_normalize_vector(list1):
    m = mean(list1)
    v = math.sqrt(variance(list1))
    return [ (value - m) / v for value in list1]

def dotProduct(list1 , list2):
    if len(list1) != len(list2):
        print("invalid input")
    else:
        result = []
        for i in range(len(list1)):
            result.append(list1[i] * list2[i])
        return sum(result)

def magnitude(list1):
    return math.sqrt( sum([math.pow(value , 2) for value in list1]))

print(addVectors([5,3,4] , [1,2,6]))
print(subtractVectors([5,3,4] , [1,2,6]))
print(scaleVector([5,3,4] , 3))
print(normalizeVector([5,3,4]))
print(z_score_normalize_vector([5,3,4]))
print(dotProduct([5,3,4] , [1,2,6]))
print(magnitude([5,3,4]))
```

```python
# TASK:5
def displayMatrix(array):
    row = len(array)
    col = len(array[0])
    for i in range(row):
        for j in range(col):
            print(array[i][j],end=" ")
        print()

def getColumn(array  ,column_index):
    result = []
    for i in range(len(array)):
        result.append(int(array[i][column_index]))
    return result

def getRow(array , row_index):
    return array[row_index]

def getMainDiagonal(matrix ):
   return [matrix[i][i] for i in range(len(matrix))]


def addMatrices(matrix1 , matrix2):
    return [[matrix1[i][j] + matrix2[i][j]  for j in range(len(matrix1[0]))] for i in range(len(matrix1))]

def subtractMatrices(matrix1 , matrix2):
    return [[matrix1[i][j] - matrix2[i][j]  for j in range(len(matrix1[0]))] for i in range(len(matrix1))]

def scaleMatrix(matrix , scale_value):
    return [[matrix[i][j] * scale_value  for j in range(len(matrix[0]))] for i in range(len(matrix))]

def transposeMatrix(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

def traceMatrix(matrix):
    return sum(getMainDiagonal(matrix))

def matrixToVector(matrix):
    return sum(matrix , [])

matrix = [['1', '2', '3', '4', '5'],
 ['6', '7', '8', '9', '1'],
 ['3', '2', '1', '4', '5']]

diagnol_matrix = [[1,2,3],
                  [4,5,6],
                  [7,8,9]]

new_matrix = [[1,2,3],
                  [4,5,6],
                  [7,8,9]]



displayMatrix(matrix)


print(getColumn(matrix , 2))

print(getRow(matrix , 2))

print(getMainDiagonal(diagnol_matrix))

print(addMatrices(diagnol_matrix , new_matrix))

print(subtractMatrices(diagnol_matrix , new_matrix))

print(scaleMatrix(diagnol_matrix  , 3))

print(transposeMatrix(diagnol_matrix))

print(traceMatrix(diagnol_matrix))

print(matrixToVector(diagnol_matrix))
```

```python

#task 7

from dataclasses import dataclass
import random
box_types=["Standard","Reinforced","Heavy Duty"]

material_types={}
material_types["Standard"]=["Thick Paper","Cardboard"]
material_types["Reinforced"]=["Taped Cardboard","Plastic"]
material_types["Heavy Duty"]=["Heavy Plastic","Metal"]

price_multipliers={}
price_multipliers['Standard']=[1.15,1.25]
price_multipliers['Reinforced']=[1.50,2.10]
price_multipliers['Heavy Duty']=[3.0,4.0]

water_light_map={'Standard':False,'Reinforced':False,'Heavy Duty':True}

water_resistant_map={'Standard':False,'Reinforced':True,'Heavy Duty':True}

color={"Standard":["Brown","Olive"],"Reinforced":["Black","Green"],"Heavy Duty":["Black","Yellow","Tan"]}

lengths={"Standard":range(10,13),"Reinforced":range(12,15),"Heavy Duty":range(16,25)}

widths={"Standard":range(6,9),"Reinforced":range(8,11),"Heavy Duty":range(12,25)}

heights={"Standard":range(8,11),"Reinforced":range(8,11),"Heavy Duty":range(12,25)}



@dataclass
class ShippingBox:
    serialNumber:int
    boxType:str
    length:float
    Width:float
    height:float
    area:float
    material:str
    color:str
    cost:float
    water_tight:bool
    water_resistant:bool

    def __init__(self, serialNumber):
        #check if starts with B or add B 
        serialNumber=str(serialNumber)
        if serialNumber[0] != "B":
            self.serialNumber = "B" + serialNumber
        else:
            self.serialNumber = serialNumber
        #select a random box type
        self.boxType=random.choice(box_types)
        self.length = random.choice(lengths[self.boxType])
        self.Width = random.choice(widths[self.boxType])
        self.height = random.choice(heights[self.boxType])
        self.material = random.choice(material_types[self.boxType])
        self.color = random.choice(color[self.boxType])
        self.water_tight = water_light_map[self.boxType]
        self.water_resistant = water_resistant_map[self.boxType]
        self.area = round(self.length * self.Width* self.height,2)
        self.cost=round(self.area*price_multipliers[self.boxType][material_types[self.boxType].index(self.material)],2)

```

```python
#task 9

def analyzeBoxes(filenmae):
    with open(filenmae, "r") as shipBoxFile:
        lines=shipBoxFile.readlines()
        #create a list of tuples
        color_distribution={}
        material_distribution={}
        shipping_box_type_distribution={}
        costs=[]
        for line in lines:
            line=line.strip()
            tokens=line.split(",")
            #print(tokens)
            shipping_box_type_distribution[tokens[1]]=shipping_box_type_distribution.get(tokens[1],0)+1
            material_distribution[tokens[6]]=material_distribution.get(tokens[6],0)+1
            color_distribution[tokens[7]]=color_distribution.get(tokens[7],0)+1
            costs.append(float(tokens[8]))

        print("Shipping Box Type Distribution")
        for key,value in shipping_box_type_distribution.items():
            print(key,":",value,end=" ")
        print()
        print("Material Distribution")
        for key,value in material_distribution.items():
            print(key,":",value,end=" ")
        print()
        print("Color Distribution")
        for key,value in color_distribution.items():
            print(key,":",value,end=" ")
        print()
        print("Cost Distribution")
        print("Total Cost:",sum(costs),end=' ')
        print("Average Cost:",sum(costs)/len(costs),end=' ')
        print("Max Cost:",max(costs),end=' ')
analyzeBoxes("ShippingBoxes.txt")
```
